```python
import os
from spleeter.separator import Separator
import shutil 
print(shutil.which("ffmpeg"))
import ffmpeg
# help(ffmpeg) 
```

```python
separator = Separator('spleeter:2stems')
input_dir = "/kaggle/input/compettioon-mp3/test/test"
output_dir = 'test_separate'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
cnt=0
for filename in os.listdir(input_dir):
    if filename.endswith('.mp3'):
        input_path = os.path.join(input_dir, filename)
        output_subdir = os.path.join(output_dir, f'vocal{os.path.splitext(filename)[0]}')
        final_output_path = os.path.join(output_dir, f'vocal{os.path.splitext(filename)[0]}.mp3')
        separator.separate_to_file(input_path, output_subdir)
        wav_path = os.path.join(output_subdir, os.path.splitext(filename)[0], 'vocals.wav')
        shutil.move(wav_path, final_output_path)
        shutil.rmtree(output_subdir)
        cnt+=1
        print(f"{cnt} files processed, remaining: {len(os.listdir(input_dir))-cnt} files.")
print("All files processed.")
```

```python
import zipfile
import os

# Path to the directory containing the files you want to download
folder_path = '/kaggle/working/test_separate'

# Path to the zip file you want to create (this will save in the working directory)
zip_path = '/kaggle/working/test_separate.zip'

# Creating a zip file
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        zipf.write(file_path, arcname=file)  # arcname strips the provided path and saves only the file

print("Zip file created!")
```
